### Schritt 1: Zeichnen der geometrischen Figuren

Die einfachsten Elemente für unsere Zeichung sind die geometrischen Formen *Dreieck* und *Rechteck*. Die einfachste Implementation sieht wie folgt aus:

```python
def draw_rectangle(height, width):
        
    turtle.reset()
    turtle.penDown()
    turtle.forward(height)
    turtle.turnRight(90)
    turtle.forward(width)
    turtle.turnRight(90)
    turtle.forward(height)
    turtle.turnRight(90)
    turtle.forward(width)
    

    # fill
    turtle.penUp();
    turtle.turnRight(120)    
    turtle.forward(2)
    turtle.penColor('GREY')
    turtle.fill()
    
def draw_triangle(length):
        
    turtle.reset()
    turtle.penDown()
    turtle.turnRight(90)
    turtle.forward(length)
    turtle.turnLeft(120)
    turtle.forward(length)
    turtle.turnLeft(120)
    turtle.forward(length)
    turtle.turnLeft(120)       

    
    # fill
    turtle.penUp();
    turtle.turnLeft(30)
    turtle.forward(4)
    turtle.penColor('RED');
    turtle.fill();

    
turtle.clear()
turtle.reset()
#draw_rectangle(100, 100)
draw_triangle(100)
turtle.display()
```
